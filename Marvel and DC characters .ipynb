```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
marvel=pd.read_csv('marvel-wikia-data.csv')
dc=pd.read_csv('dc-wikia-data.csv')
```

```python
marvel.head()
```

```python
dc.head()
```

```python
marvel_sex_count=marvel['SEX'].value_counts()
dc_sex_counts=dc['SEX'].value_counts()
```

```python
plt.figure(figsize=(15,5))
plt.bar(x=marvel_sex_count.index,height=marvel_sex_count.values,color='red')
plt.title('Marvel Gender comparision')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.bar(x=dc_sex_counts.index,height=dc_sex_counts.values,color='blue')
plt.title('DC Gender Comparision')
plt.show()
```

```python
marvel_top10=marvel.sort_values(by='APPEARANCES',ascending=False).head(10)
```

```python
plt.figure(figsize=(25,10))
plt.bar(x=marvel_top10['name'],height=marvel_top10['APPEARANCES'],color='red')
plt.xticks(rotation=90)
plt.show()
```
